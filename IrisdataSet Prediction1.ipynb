```python

#Import Library

import matplotlib.pyplot as plt
import numpy as np
from numpy import genfromtxt
from sklearn import linear_model
import pandas as pd
```

```python
#Read Csv file

file = genfromtxt("iris.data.csv", delimiter=",",dtype="str")
print(file)
```

```python
#convert categorical data in to numerical data
dic = {}
count = 0
for val in file:
    if val[4] not in dic:
        dic[val[4]] = count
        count +=1
print(dic)      

       
for val in file:
    if val[4] in dic:
        val[4] = dic[val[4]]
print(file[0]) 
```

```python

#arrang Iris-setosa,iris versicolor and Iris-virginica in alternate position.

dic1 = 0
dic2 = 50
dic3 = 100
dataSet = []
for val in range(0,50):
    dataSet.append(file[dic1])
    dataSet.append(file[dic2])
    dataSet.append(file[dic3])
    dic1 +=1
    dic2 +=1
    dic3 +=1
print(dataSet) 
```

```python

#Split dataset in TrainingSet and TestingSet.

trainingSet = file[:130]
trainingX = trainingSet[:,[0,1,2,3]]
trainingX = trainingX.astype(float )
trainingY = trainingSet[:,[4]]
trainingY = trainingY.astype(float)

testingSet = file[130:]
testingX = testingSet[:,[0,1,2,3]]
testingX = testingX.astype(float)
testingY = testingSet[:,[4]]
testingY = testingY.astype(float)

print(trainingSet)
```

```python
print(testingSet)
```

```python

#fit the TrainingSet and TestingSet in Algorithm..


lr = linear_model.LogisticRegression()
lr.fit(trainingX, trainingY)
```

```python

#print the prediction

print("predicted value is" + str(lr.predict([testingX[0]])))
print("Real value is" + str(testingY[0]))
```

```python

#Accuracy...

print("acuracy is " + str(lr.score(testingX, testingY)*100))
print(dic)
```
